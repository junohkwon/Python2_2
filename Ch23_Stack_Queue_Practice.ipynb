```python
### towers of hanoi 재귀코드.

def towersOfHanoi(n,x,y,z):
    if(n>0):
        towersOfHanoi(n-1,x,z,y)
        print('Move the top disk from tower ',x,' to top of tower ',y)
        towersOfHanoi(n-1,z,y,x)

towersOfHanoi(3,'x','y','z')
```

```python
#### stack을 사용한 towers of hanoi 코드

def toh_stack(n, x, y, z):
    global Towers
    if (n > 0):
        toh_stack(n-1, x, z, y)
        moving = Towers[x].pop()
        Towers[y].append(moving)
        print("Move the top disk("+str(moving)+ ") from tower " + x + " to top of tower " + y)
        for k, v in Towers.items():
            print('\tTower ' + k + ':', v)
        print()
        toh_stack(n-1, z, y, x)

Towers = {'A':[], 'B':[], 'C':[]}

n = 5
for i in range(n,0,-1):
    Towers['A'].append(i)

toh_stack(len(Towers['A']), 'A', 'C', 'B')
```
