```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.neural_network import MLPClassifier

import warnings
warnings.filterwarnings('ignore')

originData = pd.read_csv('C:\Project\data\creditcard.csv', index_col='Time')
originData.head()
```

## 1. Linear Regression

#### 이전 Matplotlib 과제에서 유의차가 있는것으로 추정된 6개 항목에 대해 Linear Regression 분석 
#### 대상 항목 - ['V17','V11'], ['V17','V14'], ['V1','V4'], ['V1','V3'], ['V10','V4'], ['V14','V4'] 

```python
## Regression 결과를 출력할 함수
def showResult(regr):
    print('Coefficients: ', *regr.coef_)
    print('Intercept: ', *regr.intercept_)
    print('MSE: ', np.mean((regr.predict(x_test)-y_test))**2)
    print('Score: ',regr.score(x_test,y_test))

## 대상선정
items = [['V17','V11'], ['V17','V14'], ['V1','V4'], ['V1','V3'], ['V10','V4'], ['V14','V4']]
regr = linear_model.LinearRegression()

## 각 항목별로 분석결과 출력
for item in items:
    for num in range(0,2,1):
        
        ## 동일한 Class 기준으로 데이터 분리
        x = originData[originData['Class']==num][item[0]].values
        y = originData[originData['Class']==num][item[1]].values

        x = x.reshape(len(x),1)
        y = y.reshape(len(y),1)

        ## Train과 Test 데이터를 7:3으로 분리
        x_length = len(x)
        test_count = int(len(x)*.3)

        x_train = x[:-test_count]
        x_test = x[test_count:]

        y_train = y[:-test_count]
        y_test = y[test_count:]

        ## Linear Regression Fit
        regr.fit(x_train,y_train)

        ## Chart Draw
        fig, ax = plt.subplots()
        plt.scatter(x_test,y_test)
        plt.plot(x_test,regr.predict(x_test), color='black', linewidth=3)
        ax.set_xticks(())
        ax.set_yticks(())
        ax.set_xlabel(item[0])
        ax.set_ylabel(item[1])
        plt.title('regression plot : class ' + str(num))

        showResult(regr)
        plt.show()
```

#### Class 0 데이터는 Score가 매우 낮거나 음수이므로, 비선형 관계로 추정됨
#### Class 1 데이터 기준으로 대부분의 인자가 선형관계이나,  V1과 V3 인자가 Score가 가장 높음

## 3. NuralNet

#### 전체 인자에 대해 MLPClassifier 학습 후 Test Data Set으로 Score 확인

```python
datacols = originData.columns

## 원본에서 x, y 데이터 분리
x_data = pd.DataFrame(data=originData, columns=datacols)
y_data = pd.DataFrame(data=originData, columns=['Class'])

row_length = y_data.count()
sample_length = int(row_length*0.7)

## 7:3 비율로 Train/Test 데이터 분리
x_data_train = x_data.iloc[:sample_length]
x_data_test = x_data.iloc[sample_length:]

y_data_train = y_data.iloc[:sample_length]
y_data_test = y_data.iloc[sample_length:]

## Train 데이터 Set으로 학습
mlp = MLPClassifier(hidden_layer_sizes=(50,), max_iter=10, alpha=1e-4, solver='sgd',
                   verbose=10, tol=1e-4, random_state=10, learning_rate_init=.1)

mlp.fit(x_data_train, y_data_train)

## Test 데이터 Set으로 결과 확인
print('Training set score : ', mlp.score(x_data_train, y_data_train))
print('Test set score : ',mlp.score(x_data_test, y_data_test))
```

#### 위에서 선정된 일부 인자에 대해 MLPClassifier 학습 후 Test Data Set으로 Score 확인

```python
datacols = ['V17','V11','V14','V1','V4','V3','V10']

## 원본에서 x, y 데이터 분리
x_data = pd.DataFrame(data=originData, columns=datacols)
y_data = pd.DataFrame(data=originData, columns=['Class'])

row_length = y_data.count()
sample_length = int(row_length*0.7)

## 7:3 비율로 Train/Test 데이터 분리
x_data_train = x_data.iloc[:sample_length]
x_data_test = x_data.iloc[sample_length:]

y_data_train = y_data.iloc[:sample_length]
y_data_test = y_data.iloc[sample_length:]

## Train 데이터 Set으로 학습
mlp = MLPClassifier(hidden_layer_sizes=(50,), max_iter=10, alpha=1e-4, solver='sgd',
                   verbose=10, tol=1e-4, random_state=1, learning_rate_init=.1)

mlp.fit(x_data_train, y_data_train)

## Test 데이터 Set으로 결과 확인
print('Training set score : ', mlp.score(x_data_train, y_data_train))
print('Test set score : ',mlp.score(x_data_test, y_data_test))
```

#### 학습된 로직에 대해 Test Data Set을 대입하는 경우 0.99 score 확인
#### Class0으로 Bias된 데이터라서 Train Data Set과 Test Data Set의 데이터가 유사하여
#### Test Data Set의 예측 결과가 높은것으로 판단됨
