```python
class Stack:    
    def __init__(self):
        self.items = []    
    def push(self,item):
        self.items.append(item)        
    def pop(self):
        idx = self.items[-1]
        self.items = self.items[:-1]
        return idx
    def isempty(self):
        if len(self.items) == 0:
            return True
        else:
            return False

def paren_v1(lst):    
    par_stack = Stack()
     
    for idx in range(len(lst)):     
        if lst[idx] == '(':
            par_stack.push(idx)            
        elif lst[idx] == ')':
            right_paranthesis_idx = idx            
            if not par_stack.isempty():
                left_paranthesis_idx = par_stack.pop()
                print('(%d,%d) match'%( left_paranthesis_idx , right_paranthesis_idx)) 
            else:
                print('%d no match (right)' % right_paranthesis_idx)
            
    if not par_stack.isempty():
        print('no match (left) : ', par_stack.items)
        
paren_v1('((1+(2*3)+3))))(((')        
```

```python
# (Version 2): using python's queue module (LifoQueue) 
import queue

def paren_v2(lst):

    par_stack = queue.LifoQueue()
    
    for idx in range(len(lst)):     
        if lst[idx] == '(':
            par_stack.put(idx)            
        elif lst[idx] == ')':                        
            right_paranthesis_idx = idx            
            if par_stack.qsize() > 0:
                left_paranthesis_idx = par_stack.get()
                print('(%d,%d) match'%( left_paranthesis_idx , right_paranthesis_idx)) 
            else:
                print('%d no match (right)' % right_paranthesis_idx)
                
    if par_stack.qsize() > 0:
        items = [];
        while par_stack.qsize()>0:
            items.append(par_stack.get())
        
        print('no match (left) : ', items)
            
    
paren_v2('((1+(2*3)+3))))(((')            
```
